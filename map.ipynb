```python
import pandas as pd

tabela = pd.read_csv("Cidades.csv")
display(tabela)
```

```python
import plotly.express as px

grafico = px.density_map(tabela, lon="geolocation_lng", lat="geolocation_lat", z="quantidade", map_style="open-street-map", zoom=2, radius=10)
grafico.update_layout(margin={"r":0, "t": 0, "b": 0, "l": 0})
grafico.show()
```
